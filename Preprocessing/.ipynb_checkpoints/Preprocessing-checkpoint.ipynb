```python
train_pre = gabung.iloc[0:9093]
test_pre = gabung.iloc[9094:10364]
```

Drop NA in rating

```python
train_pre.isna().sum()
```

```python
train_pre=train_pre.dropna()
```

standardisasi

```python
train_pre
```

```python
num=['Reviews','Price','Rating']
```

```python
from sklearn.preprocessing import RobustScaler
sc = RobustScaler()
train_pre[num] = sc.fit_transform(train_pre[num])
test_pre[num] = sc.transform(test_pre[num])
```

### Hapus Outlier Price

```python
#hapus Outlier
q3 = train_pre.Price.quantile(0.75)
batas = q3 + 1.5*(q3 - train_pre.Price.quantile(0.25))
train_pre = train_pre[~(train_pre['Price']>=batas)]
```

### Jadiin X y yeay
