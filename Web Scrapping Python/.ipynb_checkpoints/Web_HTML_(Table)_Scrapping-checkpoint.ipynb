# Web HTML (Table) Scrapping



misal kita ingin mendapatkan tabel dari website https://www.basketball-reference.com/leagues/NBA_2022_per_game.html

```python
# Buat frame atau tempat kita agar bisa mengkostumisasi tahun
year = '2022'
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
```

```python
# Gabungkan url + string tahun 
url = url_link.format(year)
url
```

```python
years = [2015,2016,2017,2018,2019,2020,2021,2022]
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
urls=[]
for year in years:
  url = url_link.format(year)
  urls.append(url)
  print(url)
```

```python
print(urls)
```

misal kita ingin mendapatkan tabel dari data 2018 (indeks ke 4)

```python
import pandas as pd
```

```python
df = pd.read_html(urls[4], header = 0)
df
```

kita jadikan tabel yang kita baca tadi sebagai dataframe df2018

```python
df2018=df[0]
```

Identifikasi Duplicate

```python
df2018[df2018.Player == 'Player']
```

Remove all duplicate

```python
df = df2018.drop(df2018[df2018.Player == 'Player'].index)
```

```python
df
```

```python
sns.distplot(df.Age,
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2),
             color='#FF4500')
```
