```python
# Import library
import numpy as np
from tensorflow.keras.models import Sequential,Model,load_model
from tensorflow.keras.layers import LSTM
from tensorflow.keras.optimizers import SGD,Adam
from tensorflow.keras.layers import Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
from datetime import datetime
from tensorflow.random import set_seed
from tensorflow.keras.metrics import MeanAbsoluteError,MeanSquaredError,Accuracy,Precision,Recall
```

```python
#yahoo finance as data source
#pip install yfinance
import yfinance as yf
```

```python
#See the yahoo finance ticker for your stock symbol
stock_symbol = 'BTC-USD' 
```

```python
#last 5 years data with interval of 1 day
data = yf.download(tickers=stock_symbol,period='5y',interval='1d')
```

```python
type(data)
```

```python
# Check data 
data
```

```python
# len data
len(data)
```

```python
# subsetting data / choose only open
opn = data[['Open']]
```

```python
# Make plot
opn.plot(title="Data Harga Bitcoin-USD")
plt.savefig("Data Harga Bitcoin-USD.png")
```

```python
# Make Numpy Format
ds = opn.values
```

```python
# Check Value
ds
```

```python
# Plot
plt.plot(ds)
```

```python
# Standarisasi using MinMaxScaler
normalizer=MinMaxScaler(feature_range=(0,1))
ds_scaled=normalizer.fit_transform(np.array(ds).reshape(-1,1))
```

```python
# Check length ds
len(ds), len(ds_scaled)
```

```python
# Defining test and train data sizes
train_size = int(len(ds_scaled)*0.75)
test_size=len(ds_scaled)-train_size
```

```python
# Check the len train and test
train_size,test_size
```

```python
# Set seed and random seed
np.random.seed(123)
set_seed(123)
```

```python
# Splitting data
ds_train, ds_test = ds_scaled[0:train_size,:], ds_scaled[train_size:len(ds_scaled),:]
```

```python
# len ds_train and ds_test
len(ds_train), len(ds_test)
```

```python
#creating dataset in time series for LSTM model 
def create_ds(dataset,step):
    Xtrain, Ytrain = [], []
    for i in range(len(dataset)-step-1):
        a = dataset[i:(i+step), 0]
        Xtrain.append(a)
        Ytrain.append(dataset[i + step, 0])
    return np.array(Xtrain), np.array(Ytrain)
```

```python
#Taking 100 days price as one record for training
time_stamp = 100
X_train, y_train = create_ds(ds_train,time_stamp)
X_test, y_test = create_ds(ds_test,time_stamp)
```

```python
X_train.shape,y_train.shape
```

```python
#Reshaping data to fit into LSTM model
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
# Check shape
X_train.shape, X_test.shape
```

```python
#Creating LSTM model using keras
model = Sequential()
model.add(LSTM(units=50,activation='relu',return_sequences=True,input_shape=(X_train.shape[1],1)))
model.add(LSTM(units=50,return_sequences=False,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer=Adam(learning_rate=1e-3), loss='mse',metrics=[MeanAbsoluteError(),MeanSquaredError()])
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model,"Model LSTM Univariate.png", show_shapes= True)
```

```python
#Training model with adam optimizer and mean squared error loss function
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,batch_size=1)
```

```python
#Simpan model
model.save("LSTM Univariate.h5")
```

```python
#Memuat (load) model 
model=load_model("LSTM Univariate.h5")
```

```python
# Evaluate Model
test_eval=model.evaluate(X_test, y_test)
for name, value in zip(model.metrics_names, test_eval):
  print(name, ': ', value)
print()
```

```python
%matplotlib inline

import matplotlib.image  as mpimg
import matplotlib.pyplot as plt

loss=history.history['loss']
val_loss=history.history['val_loss']
mae=history.history['mean_absolute_error']
val_mae=history.history['val_mean_absolute_error']
mse=history.history['mean_squared_error']
val_mse=history.history['val_mean_squared_error']

epochs=range(len(loss))

plt.plot(epochs, loss, 'r', label = 'Training loss')
plt.plot(epochs, val_loss, 'b', label ='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.figure()

plt.plot(epochs, mae, 'r', label = 'Training MAE')
plt.plot(epochs, val_mae, 'b', label ='Validation MAE')
plt.title('Training and validation MAE')
plt.legend()
plt.figure()

plt.plot(epochs, mse, 'r', label = 'Training MSE')
plt.plot(epochs, val_mse, 'b', label ='Validation MSE')
plt.title('Training and validation MSE')
plt.legend()
plt.figure()
```

```python
#Getting the last 100 days records
fut_inp = ds_test[357:]
```

```python
# reshape data
fut_inp = fut_inp.reshape(1,-1)
```

```python
# Check
fut_inp.shape
```

```python
tmp_inp = list(fut_inp)
```

```python
#Creating list of the last 50 data
tmp_inp = tmp_inp[0].tolist()
```

```python
#Predicting next 30 days price suing the current data
#It will predict in sliding window manner (algorithm) with stride 1
lst_output=[]
n_steps=100
i=0
while(i<30):
    
    if(len(tmp_inp)>100):
        fut_inp = np.array(tmp_inp[1:])
        fut_inp=fut_inp.reshape(1,-1)
        fut_inp = fut_inp.reshape((1, n_steps, 1))
        yhat = model.predict(fut_inp, verbose=0)
        tmp_inp.extend(yhat[0].tolist())
        tmp_inp = tmp_inp[1:]
        lst_output.extend(yhat.tolist())
        i=i+1
    else:
        fut_inp = fut_inp.reshape((1, n_steps,1))
        yhat = model.predict(fut_inp, verbose=0)
        tmp_inp.extend(yhat[0].tolist())
        lst_output.extend(yhat.tolist())
        i=i+1
    

print(lst_output)
```
